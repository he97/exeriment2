```python
import os
root_directory = '/Users/he97/PycharmProjects/experiment2'
result_directory = 'outputs/eta_maskrt_refactor'
os.getcwd()
os.chdir(f'{root_directory}/{result_directory}')
```

```python
import re
os.chdir(f'{root_directory}/{result_directory}')
d = {}
for x in os.listdir():
    if not os.path.isdir(f'{root_directory}/{result_directory}/{x}'):
        continue
    os.chdir(f'{root_directory}/{result_directory}/{x}')
    if 'log_rank0.txt' in os.listdir():
        pattern = re.compile('[0-9]{1,2}\.[0-9]{1,4} \+\- [0-9]{1,2}\.[0-9]{1,4}')
        with open('log_rank0.txt') as f:
            data = f.read()
        result = pattern.findall(data)
        if len(result)>0:
            d[x] = result
        os.chdir(f'{root_directory}/{result_directory}')
d
```

```python
import pandas as pd
columns = ['OA','AA','KAPPA']
class_num = 7
columns.extend([f'class{i+1}'for i in range(class_num)])
df = pd.DataFrame(columns=columns,index=sorted(d.keys()))
for x in sorted(d.keys()):
    df.loc[x] = d[x]
df = df.transpose()
df.to_csv('result.csv')
```
